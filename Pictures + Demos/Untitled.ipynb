```julia
using Plots
fig1 = contour(-1.5:0.1:1.5, -1.5:0.1:1.5, (x, y) -> x^2 - y^2)
fig1 = plot!(fig1, sin, cos, 0, 2, lab="x^2 + y^2 == 1", linewidth=4)
fig2 = contour(-1.5:0.1:1.5, -1.5:0.1:1.5, (x, y) -> x^2 + y^2)
fig2 = plot!(fig2, sin, cos, 0, 2, lab="x^2 + y^2 == 1", linewidth=4)
fig = plot(fig1, fig2, size=(1200, 500), layout=2)
```
