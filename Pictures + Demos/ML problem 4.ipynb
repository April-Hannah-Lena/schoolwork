```julia
# 2)
f(a) = sum(@. (x * a - y)^2)
Df(a) = sum(@. 2 * x * (x * a - y))
DDf = sum(@. 2 * x^2)
a, h = 0.5, 0
f_new, f_old = f(a), Inf
while f_old - f_new > eps()                 # simple newton-method
    h = -Df(a)/DDf                      
    @show a = a + h
    f_old = f_new
    @show f_new = f(a)
end
fig = plot!(0:0.5:1, x -> a*x, linewidth=4)
```
